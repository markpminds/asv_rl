```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import numpy as np
from matplotlib.cbook import get_sample_data
import matplotlib.image as mpimg

# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Load the boat image


# Replace this with the path to your boat image
boat_image_path = get_sample_data("/Users/markprettyman/Desktop/asv_rl/src/asv/images/Corsair.png")  # Sample image from Matplotlib
boat_image = mpimg.imread(boat_image_path)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y, label="Sine Wave")
ax.legend()

# Function to add an image to the plot
def add_image(x, y, image, zoom=0.1):
    imagebox = OffsetImage(image, zoom=zoom)
    ab = AnnotationBbox(imagebox, (x, y), frameon=False)
    ax.add_artist(ab)

# Add a boat image at a specific data point
add_image(5, np.sin(5), boat_image, zoom=0.2)

# Show the plot
plt.title("Chart with Custom Boat Image")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
